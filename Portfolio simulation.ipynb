```python
K = 1
lot = 1
port = pd.read_excel('ptf_risk.xlsx')
port
```

```python
# port['rCost'] = port['rCost'] * 1000
# port['rPrice'] = port['rPrice'] * 1000
# port['rSL'] = port['SL'] / port['Price_bm']
# port['rSL'] = port['rSL'] * 1000
```

```python
port['rSL'] = port['SL'] / port['Price_bm']
BV = port['Shares'] * port['Cost']
MV = port['Shares'] * port['Price']
rMV = port['Shares'] * port['rPrice']
MV
```

```python
port['rR'] = (port['rCost'] - port['rSL'])
port['Weight'] = round(MV.div(abs(MV).sum()),3)
port['rRisk'] = -round(np.maximum(0,(port['rR'] * port['Shares'])/K),14)
port['rRAR'] = round( (port['rPrice'] - port['rCost'])/port['rR'],14)
port['rCTR'] = round(port['Shares'] * (port['rPrice']-port['rCost'])/ K,14)
port['CTR'] = round(port['Shares'] * (port['Price']-port['Cost'])/ K,14)
port_long = port[port['Side']>0]
port_short = port[port['Side']<0]
concentration = (port_long['Side'].count()-port_short['Side'].count())/port['Side'].count()
```

```python
gross = round(abs(MV).sum() / K,3) 
net = round(MV.sum()/abs(MV).sum(),3)
net_Beta = round((MV* port['Beta']).sum()/abs(MV).sum(),2)
print('Gross Exposure',gross,'Net Exposure',net,'Net Beta',net_Beta,'concentration',concentration)
```

```python
adjust_long = adjust_short  =  -0.01 

pro_rata_long = port_long['rRisk'] / (port_long['rRisk'].sum() * port_long['rRAR'])
risk_adj_long = (abs(adjust_long) * pro_rata_long * K / port_long['rR'] // lot) * lot
shares_adj_long =  np.minimum(risk_adj_long, port_long['Shares'])*np.sign(adjust_long)
shares_adj_long
```

```python
port_long.loc[:, 'pro_rata_long'] = port_long['rRisk'] / (port_long['rRisk'].sum() * port_long['rRAR'])  
port_long.loc[:, 'risk_adj_long'] = (abs(adjust_long) * port_long['pro_rata_long'] * K / port_long['rR'] // lot) * lot  
port_long.loc[:, 'shares_adj_long'] =  np.minimum(port_long['risk_adj_long'], port_long['Shares']) * np.sign(adjust_long)  
port_long
```

```python
port_short['pro_rata_short'] = port_short['rRisk'] / (port_short['rRisk'].sum() * port_short['rRAR'])
port_short['risk_adj_short'] = (abs(adjust_short) * port_short['pro_rata_short'] * K / port_short['rR'] // lot) * lot  
port_short['shares_adj_short'] =  np.maximum(port_short['risk_adj_short'], port_short['Shares'])*np.sign(adjust_short)
port_short
```

```python
port_short.loc[:, 'pro_rata_short'] = port_short['rRisk'] / (port_short['rRisk'].sum() * port_short['rRAR'])  
port_short.loc[:, 'risk_adj_short'] = (abs(adjust_short) * port_short['pro_rata_short'] * K / port_short['rR'] // lot) * lot  
port_short.loc[:, 'shares_adj_short'] =  np.maximum(port_short['risk_adj_short'], port_short['Shares']) * np.sign(adjust_short)  
port_short
```

```python
combined_shares_adj = pd.concat([port_short['shares_adj_short'], port_long['shares_adj_long']])  
port['Qty_adj'] = combined_shares_adj
port
```

```python

port['Shares_adj']  = port['Shares'] + port['Qty_adj']
port['rRisk_adj'] = -round(np.maximum(0,(port['rR'] * port['Shares_adj'])/K),4)
MV_adj= port['Shares_adj'] * port['Price']
rMV_adj = port['Shares_adj'] * port['rPrice']
port['Weight_adj'] = round(MV_adj.div(abs(MV_adj).sum()),3)

(port[['Side','rRAR','rRisk','rRisk_adj','Shares','Qty_adj', 'Shares_adj', 'Weight','Weight_adj']].groupby('Side').sum())
```
