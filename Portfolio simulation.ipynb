```python
import sys  
import os  
import pandas as pd  
sys.path.insert(0, '..')
from AT_funs import *
from datetime import datetime, date 
```

```python
K = 1
lot = 1
port = pd.read_excel('PTF.xlsx')
port['Side'] = np.sign(port['Shares'])
ticker_list = port['ticker']
# port = port.set_index('ticker')  
port['date_entry'] = pd.to_datetime(port['date_entry']) 
port
```

```python
filename_bm = 'FTSEMIB.MI.xlsx'
bm_df = read_xlsx(filename_bm)
bm_name = filename_bm.replace('xlsx', '')
bm_df['date'] = pd.to_datetime(bm_df['date'])
bm_df['date'] = bm_df['date'].dt.strftime('%Y-%m-%d')  
bm_df['date'] = pd.to_datetime(bm_df['date'])  
bm_price = bm_df['close'].iloc[-1]
bm_cost = bm_df['close'].iloc[0]
bm_df = bm_df.rename(columns={'close': 'Cost_bm'})  
bm_df
```

```python
filename = ticker_list
date_entry = port['date_entry']
date_entry = np.array(date_entry, dtype='datetime64')  
folder_path = os.getcwd()  
```

```python
price_df = []
for i in range(len(filename)):  
    f = filename[i]
    de = date_entry[i]
    df = read_xlsx(f)
    df['date'] = pd.to_datetime(df['date'])
    df['name'] = str(f)
    df = df[df['date'] >= de] 
    price_df.append(df)
price_df = pd.concat(price_df, axis=0, ignore_index=True) 
price_df['date'] = pd.to_datetime(price_df['date'])
price_df['date'] = price_df['date'].dt.strftime('%Y-%m-%d')  
price_df['date'] = pd.to_datetime(price_df['date'])  
price_df = price_df.rename(columns={'close': 'Price'}) 
```

```python
price_df
```

```python
port = pd.merge(port, bm_df[['date', 'Cost_bm']], left_on='date_entry', right_on='date', how='left')
```

```python
port = port.drop('date', axis=1)
# to be changed
port['Cost_bm'] = port['Cost_bm'].fillna(bm_price) 
#
port['Price_bm'] = bm_price
port['rCost'] = port['Cost'] / port['Cost_bm']
port
```

```python
port = pd.merge(port, price_df[['date', 'Price', 'name']], left_on=['date_entry', 'ticker'], right_on=['date', 'name'], how='left')
port = port.drop('date', axis=1)
port = port.drop('name', axis=1)
port
```

```python
port.loc[port['ticker'] == 'FCT.MI.xlsx', 'Price'] = port.loc[port['ticker'] == 'FCT.MI.xlsx', 'Price'].fillna(0.586)
port.loc[port['ticker'] == 'BGN.MI.xlsx', 'Price'] = port.loc[port['ticker'] == 'BGN.MI.xlsx', 'Price'].fillna(38.5)
```

```python
port['rPrice'] = port['Price'] / port['Price_bm']
port
```

```python
port['rSL'] = port['SL'] / port['Price_bm']
BV = port['Shares'] * port['Cost']
MV = port['Shares'] * port['Price']
rMV = port['Shares'] * port['rPrice']
MV
```

```python
port['rR'] = (port['rCost'] - port['rSL'])
port['Weight'] = round(MV.div(abs(MV).sum()),3)
port['rRisk'] = -round(np.maximum(0,(port['rR'] * port['Shares'])/K),14)
port['rRAR'] = round( (port['rPrice'] - port['rCost'])/port['rR'],1)
port['rCTR'] = round(port['Shares'] * (port['rPrice']-port['rCost'])/ K,14)
port['CTR'] = round(port['Shares'] * (port['Price']-port['Cost'])/ K,14)
port_long = port[port['Side']>0]
port_short = port[port['Side']<0]
concentration = (port_long['Side'].count()-port_short['Side'].count())/port['Side'].count()
```

```python
gross = round(abs(MV).sum() / K,3) 
net = round(MV.sum()/abs(MV).sum(),3)
net_Beta = round((MV* port['Beta']).sum()/abs(MV).sum(),2)
print('Gross Exposure',gross,'Net Exposure',net,'Net Beta',net_Beta,'concentration',concentration)
```

```python
rnet = round(rMV.sum()/abs(rMV).sum(),3)
rnet_Beta = round((rMV* port['Beta']).sum()/abs(rMV).sum(),2)
print('rGross Exposure',gross,'rNet Exposure',rnet,'rNet Beta',rnet_Beta)
```

```python
adjust_long = adjust_short  =  -0.01 
pro_rata_long = port_long['rRisk'] / (port_long['rRisk'].sum() * port_long['rRAR'])
risk_adj_long = (abs(adjust_long) * pro_rata_long * K / port_long['rR'] // lot) * lot
shares_adj_long =  np.minimum(risk_adj_long, port_long['Shares'])*np.sign(adjust_long)
shares_adj_long
```

```python
pro_rata_short = port_short['rRisk'] / (port_short['rRisk'].sum() * port_short['rRAR'])
risk_adj_short = (abs(adjust_short) * pro_rata_short * K / port_short['rR'] // lot)*lot
shares_adj_short = np.maximum(risk_adj_short,port_short['Shares'])*np.sign(adjust_short)
```

```python
shares_adj_short
```

```python
port['Qty_adj'] = shares_adj_short.append(shares_adj_long)
# port['Qty_adj'] = shares_adj_long
port['Shares_adj']  = port['Shares'] + port['Qty_adj']
port['rRisk_adj'] = -round(np.maximum(0,(port['rR'] * port['Shares_adj'])/K),4)
MV_adj= port['Shares_adj'] * port['Price']
rMV_adj = port['Shares_adj'] * port['rPrice']
port['Weight_adj'] = round(MV_adj.div(abs(MV_adj).sum()),3)

print(port[['Side','rRAR','rRisk','rRisk_adj','Shares','Qty_adj', 'Shares_adj', 'Weight','Weight_adj']].groupby('Side').sum())
```
